# Displots

```python
sns.set(rc={'figure.figsize':[10,10]},font_scale=1.2)
```

```python
sns.distplot(data['National Income'],color='blue',kde=False,bins=10)
```

```python
sns.distplot(data['Growth'],color='blue',kde=False,bins=10)
```

```python
sns.distplot(data['Per person'],color='blue',kde=False,bins=10)
```

# kdeplot

```python
sns.kdeplot(data['National Income'],shade=True,color='red')
```

```python
sns.kdeplot(data['Growth'],shade=True,color='red')
```

```python
sns.kdeplot(data['Per person'],shade=True,color='red')
```

# Bivariate Plots
## JointPlot

```python
sns.jointplot(x='Growth',y='National Income',data=data,kind='scatter',cmap='viridis',height=10,ratio=10) # No relation between 
                                                                                                         # variables
```

```python
sns.jointplot(x='Growth',y='Per person',data=data,kind='kde',height=10,ratio=10) # decreasing relationship between the per
                                                                                 # person and the growth of each country
```

```python
sns.jointplot(x='Growth',y='National Income',data=data,kind='hex',height=10,ratio=10) # No relation between Variables
```
