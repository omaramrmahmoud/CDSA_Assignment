# Task 2

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
df=pd.read_csv('ufo.csv')
df.head()
```

```python
df.info()
```

```python
df['Colors Reported'].value_counts()
```

```python
df['Shape Reported'].value_counts()
```

```python
df['Colors Reported'].fillna(df['Colors Reported'].mode()[0], inplace=True)
df['Shape Reported'].fillna(df['Shape Reported'].mode()[0], inplace=True)
df['City'].fillna(df['City'].mode()[0], inplace=True)
df.isnull().sum()
```

```python
df['Time'] = pd.to_datetime(df['Time'], format='%m/%d/%Y %H:%M', errors="coerce")
df.head()
```

```python
df['Month'] = df['Time'].dt.month
df.head()
```

```python
df['IN_December']=df['Month'].apply(lambda x: (1) if(x==12) else (0))
df.drop('Month', 1,inplace=True)
df.head()
```

```python
from datetime import datetime

ww1_date = datetime(1914, 7, 28)

df['Time_Diff_From_WW1']  = (df['Time'] - ww1_date) / np.timedelta64(1, 'Y')
df
```

# Task 6

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(rc={'figure.figsize':(10, 10)}, font_scale=1.2)

df=pd.read_csv('Restaurant_Reviews.csv',sep='\t')
df.head()
```

```python
reviews = df['Review']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(stop_words='english')
reviews_vecctor = vectorizer.fit_transform(reviews)
reviews_vecctor.toarray()
```

```python
pd.DataFrame(reviews_vecctor.toarray(), columns=vectorizer.get_feature_names())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words='english')
reviews_vecctor = vectorizer.fit_transform(reviews)
reviews_vecctor.toarray()
```

```python
pd.DataFrame(reviews_vecctor.toarray(), columns=vectorizer.get_feature_names())
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(rc={'figure.figsize':(10, 10)}, font_scale=1.2)

df=pd.read_csv('Salaries.csv')
df.head()
```

```python
df = df[['BasePay', 'OvertimePay', 'OtherPay', 'Benefits', 'TotalPay', 'TotalPayBenefits']]
df.head()
```

```python
df = df.dropna()
df.head()
```

```python
X = df[['BasePay', 'OvertimePay', 'OtherPay', 'Benefits', 'TotalPay']]
y = df['TotalPayBenefits']
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
from sklearn.decomposition import PCA

pca = PCA(0.9)

X = pca.fit_transform(X)
```

```python
pca.n_components_
```

```python
df = pd.DataFrame(data=X)
df
```
