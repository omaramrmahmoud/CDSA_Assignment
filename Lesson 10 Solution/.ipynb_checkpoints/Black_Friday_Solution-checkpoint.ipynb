```python
df = df.drop(['User_ID', 'Product_ID', 'Product_Category_3'], axis=1) # we will drop the first two columns as they are of no use
df.head()                                                             # and the third one has alot of missing data
```

```python
df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0], inplace=True) # we used first mode to impute, as it is a
                                                                                  # categorical numeric data
```

```python
df.isnull().sum()
```

```python
import datasist as ds
outliers_indices = ds.structdata.detect_outliers(df, 0, df.select_dtypes(include=['number']).columns)
df.drop(outliers_indices, inplace=True)
```

```python
df = pd.get_dummies(df, columns=df.columns[:-1], drop_first=True)
df.head()
```

```python
x = df.drop('Purchase', axis=1)
y = df['Purchase']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=22)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```
